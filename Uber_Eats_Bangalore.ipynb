```python
#name
df['name'] = df['name'].str.strip().str.title()
```

```python
#online order
df['online_order'] = df['online_order'].str.strip().str.lower()
df['online_order'] = df['online_order'].map({'yes': 1, 'no': 0})
```

```python
#book_table
df['book_table'] = df['book_table'].str.strip().str.lower()
df['book_table'] = df['book_table'].map({'yes': 1, 'no': 0})
```

```python
#votes
df['votes'] = pd.to_numeric(df['votes'], errors='coerce').fillna(0).astype(int)
```

```python
#location
df['location'] = df['location'].str.strip().str.title()
```

```python
df['approx_cost(for two people)'] = (
    df['approx_cost(for two people)']
    .astype(str)
    .str.replace(',', '', regex=False)
)

df['approx_cost(for two people)'] = pd.to_numeric(
    df['approx_cost(for two people)'],
    errors='coerce'
)
```

```python
df['listed_in(type)'] = df['listed_in(type)'].str.strip().str.title()
```

```python
df.duplicated().sum()
```

```python
df['rate'] = df['rate'].astype(str)

df['rate'] = df['rate'].replace(['NEW', '-', 'nan'], np.nan)

df['rate'] = df['rate'].str.split('/').str[0]

df['rate'] = pd.to_numeric(df['rate'], errors='coerce')
```

```python
df = df.drop_duplicates().reset_index(drop=True)
```

```python
df['rest_type'] = df['rest_type'].astype(str)

df['rest_type_primary'] = df['rest_type'].str.split(',').str[0].str.strip()
```

```python
df['cuisines'] = (
    df['cuisines']
    .astype(str)
    .str.strip()
    .str.replace(r'\s*,\s*', ', ', regex=True)
)
```

```python
unique_cuisines = (
    df['cuisines']
    .dropna()
    .str.split(', ')
    .explode()
    .unique()
)
```

```python
unique_cuisines_list = unique_cuisines.tolist()
```

```python
cuisine_temp = (
    df[['name', 'cuisines']]
    .dropna(subset=['cuisines'])
    .assign(cuisine=df['cuisines'].str.split(', '))
    .explode('cuisine')
)
```

```python
cuisine_counts = (
    cuisine_temp
    .groupby('cuisine')['name']
    .nunique()
    .reset_index(name='restaurant_count')
)
```

```python
most_represented_cuisine = cuisine_counts.sort_values(
    by='restaurant_count',
    ascending=False
).head(1)
```
